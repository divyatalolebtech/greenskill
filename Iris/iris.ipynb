```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

ec = [1200,1300,1100,1500,1400,1600]

plt.plot(months, ec, marker ='o', color='b', linestyle = '--')

plt.title('Energy Consumption Over 6 Monthys')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
plt.show()
```

```python
plt.bar(months, ec,color='g')

plt.title("Energy Consumption Over 6 Months")
plt.xlabel("Month")
plt.ylabel("Energy Consumption (MWh)")

plt.show()
```

```python
plt.pie(ec,labels=months,autopct='%1.1f%%')

plt.title("Energy Consumption Over 6 Months")

plt.show()
```

```python
plt.scatter(months,ec,color='red')
plt.title("Energy Consumption Over 6 Months")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
data = {
    "Solar": [1200,1500,1300],
    "Wind":[3400, 3600, 3200],
    "Hydropower": [2900, 3100,2800],
    "Biomass": [2500, 2700, 2400]
}
```

```python
df = pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
mat = df.corr()
sns.heatmap(mat, annot=True, cmap='coolwarm',linewidths=0.5)
plt.title("Correlation Between Energy Sources")
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
sns.violinplot(data=df)
plt.show()
```

```python
ev = [100,200,300,400,500]
ce = [10,20,25,40,50]

dfr = pd.DataFrame({'Energy Consumption' : ev,
                    'Carbon Emissions' : ce})
sns.regplot(x = 'Energy Consumption',y = 'Carbon Emissions', data = dfr)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig=plt.figure()
ax=plt.axes(projection='3d')
z = np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```

```python
fp = "IRIS.csv"
df = pd.read_csv(fp)
```

```python
df.head()
```

```python
sns.pairplot(df, hue='species', palette='bright')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
correlation_matrix = df.iloc[:, :-1].corr()  # Exclude 'species' column for correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Feature Correlation')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```

```python

sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violinplot of Petal Length by Species')
plt.show()
```
